```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline
```

```python
df = pd.read_csv("/Applications/Gursimran/Projects/Python_Hands_on/train.csv")
print(df.shape)
df.head()
```

```python
x = np.arange(-5, 5, 0.1)
y = np.square(x)  #squared x
plt.plot(x, y)    #plotting it
plt.title('$y=x^2$')   #assigning the title for the graph with correct sign
plt.xlabel('x')        #assigning the x axis with a label 
plt.ylabel('y')        #assigning the y axis with a label
plt.show()             #to get the graph window
```

```python
y2 = x ** 3    #calculating the y value
plt.figure(figsize=(12, 5))     #fixing the figure size
plt.plot(x, y, label="$y=x^2$") #legends help in labels
plt.plot(x, y2, label="y=x^3")  
plt.title('powers of x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()   #shows the legends in left top corner
plt.grid()     #shows the grid
plt.show()
plt.savefig("figure1.png")  #to save the figure
```

```python
plt.hist(df.Fare, bins=30)    # can assign the bins to please the scale
plt.grid()
plt.xlabel("Fare")
plt.title("Fare Histogram")
plt.show()
```

```python
df.Pclass.unique()
```

```python
df['log_fare'] = np.log(1 + df.Fare)
for c in df.Pclass.unique():
    plt.hist(df.loc[df.Pclass==c, 'log_fare'], density = True, bins=30, alpha=0.3, label=c)
plt.legend()
plt.show()
#normalizing using density
```

```python
plt.plot(df.Age, df.Fare)
plt.title('Fare vs Age')
plt.show()
#this is not we want, it is not clear
```

```python
plt.scatter(df.Age, df.Fare)
plt.title('Fare vs Age')
plt.show()
#now this is making some sense
```

```python
df.dropna(inplace=True, subset=["Fare", "Age"])
df.shape
```

```python
from sklearn.linear_model import LinearRegression
 
model = LinearRegression()
model.fit(df.Age.values.reshape(-1, 1), df.Fare.values.reshape(-1, 1))
x = np.arange(df.Age.min(), df.Age.max(), 0.1).reshape(-1, 1)
y = model.predict(x)
```

```python
plt.scatter(df.Age, df.Fare, label="data")
plt.scatter(x, y, label="best fit")
plt.title('Fare vs Age')
plt.legend()
plt.show()
```

```python
plt.scatter(df.Age, df.Fare, label="data", c="c")
plt.plot(x, y, label="best fit", c="r")
plt.title('Fare vs Age')
plt.legend()
plt.show()
#two plots on top of each other
```
