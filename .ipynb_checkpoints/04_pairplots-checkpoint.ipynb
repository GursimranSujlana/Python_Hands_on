```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
```

```python
data = load_iris()
data
```

#### sklearn gives is a dictionary and we need to convert it into dataframe

```python
df = pd.DataFrame(data["data"], columns=data["feature_names"])
df["target"] = [data["target_names"][i] for i in data["target"]]
#done to separate the names
df
# 4 features and 1 target
```

#### seaborn plots are better looking than matplotlib. It is using matplotlib under neadth

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, vars=df.columns[:2])
```

```python
sns.pairplot(df, vars=df.columns[:2], hue="target")
```
